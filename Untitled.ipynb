```python
mask = np.load('C:/Users/archy/Desktop/Учёба/Сессия/Курсач/Research/data/test_fold_1/test_borders.npy') #np.load('data/train/seismic.npy')
def create_borders(mask, null=False):
    if null:
        new_mask = mask.copy() * 0
    else:
        new_mask = mask.copy()
        
    for i in range(new_mask.shape[0]):
        for y in range(0, new_mask.shape[1]):
            for x in range(0, new_mask.shape[2]):
                if x != 0 and x != new_mask.shape[2] - 1:
                    if mask[i][y][x] != mask[i][y][x + 1]:
                        new_mask[i][y][x] = 1
                    if mask[i][y][x] != mask[i][y][x - 1]:
                         new_mask[i][y][x] = 1
                elif x == 0:
                    if mask[i][y][x] != mask[i][y][x + 1]:
                        new_mask[i][y][x] = 1
                elif x == new_mask.shape[2] - 1:
                    if mask[i][y][x] != mask[i][y][x - 1]:
                        new_mask[i][y][x] = 1
                        
                if y != 0 and y != new_mask.shape[1] - 1:
                    if mask[i][y][x] != mask[i][y + 1][x]:
                        new_mask[i][y][x] = 1
                    if mask[i][y][x] != mask[i][y - 1][x]:
                         new_mask[i][y][x] = 1
                elif y == 0:
                    if mask[i][y][x] != mask[i][y + 1][x]:
                        new_mask[i][y][x] = 1
                elif y == new_mask.shape[1] - 1:
                    if mask[i][y][x] != mask[i][y - 1][x]:
                        new_mask[i][y][x] = 1                    
    return new_mask

#d = []
#for i in range(0, 3):
    #d.append(np.sum(mask==0) / 35751000)
    #mask = create_borders(mask)
#print(d)
```

```python
from modules.augmentations import InvertImg
import albumentations
"""
    Augmentations:
    0. RandomCrop
    1. InvertImg
    2. Horizontal Flip
    3. CoarseDropout
    4. Blur
    5. ShiftScaleRotate [WARN] Dont rotate > 5 grad
    6. RandomBrightness
    7. All augs
"""
aug = lambda im, augm: augm(image=im)['image']
img = aug(seismic[1].T, crp)

crp = albumentations.RandomCrop(128, 256)
inv = InvertImg(p=1.0)
ct = albumentations.Cutout(num_holes=32, max_h_size=4, max_w_size=4, p=1.0)
bc = albumentations.HorizontalFlip(p=1)
blr = albumentations.GaussianBlur(6, p=1.0)
scr = albumentations.ShiftScaleRotate(rotate_limit=(-5, 5), p=1.0, border_mode=0)
rb = albumentations.Downscale(p=1.0, scale_min=0.60, scale_max=0.85)

img_all = img.copy()
img_all = aug(img_all, inv)
img_all = aug(img_all, ct)
img_all = aug(img_all, bc)
img_all = aug(img_all, blr)
img_all = aug(img_all, scr)
img_all = aug(img_all, rb)

fig, ax = plt.subplots(nrows=2, ncols=4, figsize=(16, 6))

ax[0][0].imshow(img, cmap='seismic')
ax[0][0].set_title('Random crop 64x128')

ax[0][1].imshow(aug(img, inv), cmap='seismic')
ax[0][1].set_title('InvertImg')

ax[0][2].imshow(aug(img.astype(np.float32), bc), cmap='seismic')
ax[0][2].set_title('Horizontal flip')

ax[0][3].imshow(aug(img, ct), cmap='seismic')
ax[0][3].set_title('CutOut / CoarseDropout')

ax[1][0].imshow(aug(img, blr), cmap='seismic')
ax[1][0].set_title('Gaussian Blur')

ax[1][1].imshow(aug(img, scr), cmap='seismic')
ax[1][1].set_title('ShiftScaleRotate')

ax[1][2].imshow(aug(img.astype(np.float32), rb), cmap='seismic')
ax[1][2].set_title('Downscale')

ax[1][3].imshow(img_all, cmap='seismic')
ax[1][3].set_title('All augmentations')
```
